# Energy in NVT

```python
# read the NVT energy from ener file
nvt_energy = np.loadtxt('Outputs/NVT-1.ener', skiprows=1)
# plot the potential energy vs time
plt.plot(nvt_energy[:,1], nvt_energy[:,4])
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs Time in NVT Ensemble')
plt.savefig('Figures/NVT-1_potential_energy.pdf', bbox_inches='tight')
plt.show()
```
