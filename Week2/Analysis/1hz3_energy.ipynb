```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
os.chdir('/home/ch3859/comp-class/comp-lab-class/Week2/Analysis/')
```

```python
# reads in the data file Data/1hz3_T310.run.25000000.energy.xvg
energy_data = np.loadtxt('../Data/1hz3_T310.run.25000000.energy.xvg', skiprows=29)
energy_data.shape
```

Make 6 separate plots for each quantity vs time, with time in ns

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,1])  # convert ps to ns
plt.xlabel('Time (ns)')
plt.ylabel('Potential (kJ/mol)')
plt.title('Potential Energy vs time')
plt.savefig('../Figures/1hz3_potential_energy_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,2])
plt.xlabel('Time (ns)')
plt.ylabel('Kinetic En. (kJ/mol)')
plt.title('Kinetic Energy vs time')
plt.savefig('../Figures/1hz3_kinetic_energy_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,3])
plt.xlabel('Time (ns)')
plt.ylabel('Total Energy (kJ/mol)')
plt.title('Total Energy vs time')
plt.savefig('../Figures/1hz3_total_energy_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,4])
plt.xlabel('Time (ns)')
plt.ylabel('Temperature (K)')
plt.title('Temperature vs time')
plt.savefig('../Figures/1hz3_temperature_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,5])
plt.xlabel('Time (ns)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure vs time')
plt.savefig('../Figures/1hz3_pressure_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
plt.plot(energy_data[:,0] * 0.001, energy_data[:,6])
plt.xlabel('Time (ns)')
plt.ylabel('Volume (nm^3)')
plt.title('Volume vs time')
plt.savefig('../Figures/1hz3_volume_vs_time.pdf', bbox_inches="tight")
plt.show()
```

```python
# Make a plot that has kinetic energy and temperature on the same plot, where the left axis is temperature and the right axis is kinetic energy, using different colors and different axis labels.
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(energy_data[:,0] * 0.001, energy_data[:,2], color='b', linestyle='--', marker='o', markersize=3)
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Kinetic Energy (kJ/mol)', color='b')
ax1.tick_params('y', colors='b')
ax2.plot(energy_data[:,0] * 0.001, energy_data[:,4], color='r', linestyle=':', marker='o', markersize=3)
ax2.set_ylabel('Temperature (K)', color='r')
ax2.tick_params('y', colors='r')
plt.savefig('../Figures/1hz3_kinetic_energy_temperature_vs_time.pdf', bbox_inches="tight")
plt.show()
```
